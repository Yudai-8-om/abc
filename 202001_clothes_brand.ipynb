```
from fastai import *
from fastai.vision import *
```

```
%reload_ext autoreload
%autoreload 2
%matplotlib inline
```

```
brands = {'Nike': 'nike.txt',
         'Adidas': 'adidas.txt',
         'Tommy Hilfiger': 'tommy_hilfiger.txt',
         'Ralph Lauren': 'ralph_lauren.txt',
         'Lacoste': 'lacoste.txt',
         'Gucci': 'gucci.txt',
         'Versace': 'versace.txt',
         'Under Armour': 'under_armor.txt',
         'Hollister': 'hollister.txt',
         'Levi Strauss': 'levi_strauss.txt',
         'Supreme': 'supreme.txt',
         'Burberry': 'burberry.txt',
         'Abercrombie & Fitch': 'abercrombie_fitch.txt'}
```

```
path = Path('/content/gdrive/My Drive/DL_datas/images/clothes')
for folder in brands:
  dest = path/folder
  dest.mkdir(parents=True, exist_ok=True)
```

```
path.ls()
```

```
for folder, file in brands.items():
  download_images(path/folder/file, path/folder, max_pics=1250)
```

```
for brand in brands:
  print(brand)
  verify_images(path/brand, delete=True, max_size=500)
```

```
np.random.seed(0)
data = ImageDataBunch.from_folder(path, train='.', valid_pct=0.2, ds_tfms=get_transforms(do_flip=False),
                                  size=224, num_workers=4).normalize(imagenet_stats)
```

```
data.show_batch(rows=3, figsize=(7,8))
```

```
learn = cnn_learner(data, models.resnet50, metrics=error_rate)
```

```
learn.fit_one_cycle(5)
```

```
learn.save('stage-1')
```

```
learn.unfreeze()
```

```
learn.lr_find()
```

```
  learn.recorder.plot()
```

```
learn.load('stage-1')
```

```
learn.fit_one_cycle(4, max_lr=slice(1e-5,1e-4))
```

```
learn.save('stage-2')
```

```
learn.freeze()
```

```
learn.fit_one_cycle(2)
```

```
learn.load('stage-2')
```

```
interp = ClassificationInterpretation.from_learner(learn)
```

```
interp.plot_top_losses(9, figsize=(15,11))
```

```
interp.most_confused(min_val=2)
```

```
learn.export()
```
